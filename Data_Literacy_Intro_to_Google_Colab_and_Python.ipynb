# Top Fifty Chart Busters Analysis!

## **Read in Outside Data**

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/BivinSadler/Hockaday-DSSP-MS/main/Data/TopFifty.csv'

df = pd.read_csv(url)

display(df)
```

## **Change Column Names**

```python
new_column_names = ['IndexNumber', 'Title', 'Artist','Genre','Year','BPM','Energy','Danceability','Loudness','Liveness','Valence','Length','Accousticness','Speechiness','Popularity']
df.columns = new_column_names
display(df)
```

## **Filter**

```python
# Filter Top 3
#df3 = df[(df['Genre'=='dance pop'] | df['Genre'=='pop'])]

df3 = df.query("Genre == 'pop' | Genre == 'dance pop'")
display(df3)
```

## **Make a Bar Chart of Popular Artists**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a barplot
sns.countplot(x='Artist', data=df)

# Add labels and title
plt.xlabel('Artist')
plt.ylabel('Number of Top Fifties')
plt.title('Most Frequent Artists')
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python

# Count the occurrences of each Artist
artist_counts = df['Artist'].value_counts()

# Select the top 10
top_ten_artists = artist_counts.nlargest(10)

# Create a countplot with only the top 10 Artists
sns.barplot(x=top_ten_artists.index, y=top_ten_artists.values)

# Add labels and title
plt.xlabel('Artist')
plt.ylabel('Number of Top Fifties')
plt.title('Most Frequent Artists')
plt.xticks(rotation=90)

# Show plot
plt.show()
```

## **You Try It**

1. What is the most popular (frequent) genre of song in the top fifty over the years?
2. Are there any songs that have been in the top fifty in multiple years?



```python
# Count the occurrences of each Genre
genre_counts = df['Genre'].value_counts()

# Select the top 40
top_ten_genre = genre_counts.nlargest(40)

# Create a countplot with only the top 40 genres
sns.barplot(x=top_ten_genre.index, y=top_ten_genre.values)

# Add labels and title
plt.xlabel('Genre')
plt.ylabel('Number of Top Fifties')
plt.title('Most Frequent Genres')
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
# Count the occurrences of each fruit
title_counts = df['Title'].value_counts()

# Select the top 3
top_ten_titles = title_counts.nlargest(20)

# Create a countplot with only the top 3 fruits
sns.barplot(x=top_ten_titles.index, y=top_ten_titles.values)

# Add labels and title
plt.xlabel('Titles')
plt.ylabel('Number of Top Fifties')
plt.title('Most Frequent Titles')
plt.xticks(rotation=90)

# Show plot
plt.show()
```
